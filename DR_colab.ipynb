```
task = "SST-2"

with open(f"/content/glue_data/{task}/train.tsv") as train_f:
    lines = train_f.read().splitlines()

lines_np = np.array(lines)
np.random.shuffle(lines_np[1:])

# take a percentage of shuffled dataset
lines_np = lines_np[: int(lines_np.shape[0] * (DATA_PERCENT / 100.))]

# rewrite train file with new reduced dataset
with open(f"/content/glue_data/{task}/train.tsv", "w") as train_f:
    train_f.write("\n".join(list(lines_np)))
```
